```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Suppress warnings in Python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("dataset_20220203_olympics.csv")
data.head()
```

```python
data.info()
```

```python
data.columns = data.columns.str.lower()
data.head(1)
```

```python
data = data.drop(["id", "games"], axis=1)
data
```

```python
# We will fill missing values in the height and weight columns with the activity average.
unique_activity = pd.unique(data.event)
print("Number of unique activities: {}".format(len(unique_activity)))
unique_activity[:10]
```

## Filling Missing Data in Height and Weight Columns

We will fill the missing values in the height and weight columns based on the average values for each activity.


```python
# Create a temporary copy of the dataset to avoid modifying the original data

temp_data = data.copy()  # Define the temp_data variable to preserve the original data
height_weight_columns = ["height", "weight"]

for activity in unique_activity:  # Iterate through the list of unique activities
    
    # Create a filter for the current activity
    activity_filter = temp_data.event == activity
    # Filter the data by activity
    filtered_data = temp_data[activity_filter]
    
    # Calculate the mean for height and weight for each activity
    for column in height_weight_columns:
        mean_value = np.round(np.mean(filtered_data[column]), 2)
        if ~np.isnan(mean_value):  # If there is a mean for the current activity
            filtered_data[column] = filtered_data[column].fillna(mean_value)
        else:  # If there is no mean for the current activity, calculate the overall mean
            overall_mean = np.round(np.mean(temp_data[column]), 2)
            filtered_data[column] = filtered_data[column].fillna(overall_mean)
    
    # Update the temporary dataset with the filled data for the current activity
    temp_data[activity_filter] = filtered_data
    
# Replace the original data with the filled temporary dataset
data = temp_data.copy()

# Check the number of missing values in the height and weight columns
data.info()
data.head()
```

## Filling Missing Data in Age Column

We will fill the missing values in the age column based on the average age of the dataset.


```python
# We will fill the missing values in the age column based on the average age of the dataset.
# Find the examples where the 'age' variable is not defined, 
# and use the tilde sign to get the inverse
# Create a filter to identify the examples where the 'age' variable is defined

age_average = np.round(np.mean(data.age), 2)
print("Average age: {}".format(age_average))

# Fill missing values in the 'age' column with the calculated average age
data["age"] = data["age"].fillna(age_average)

# Display the summary of the dataset
data.info()
```

## Removing Athletes Who Didn't Win Medals

We will remove athletes who did not win medals from the dataset.


```python
# We will remove athletes who did not win medals from the dataset.
# The 'medal' variable is undefined for a total of 231,333 samples.

medal_column = data["medal"]
pd.isnull(medal_column).sum()
```

```python
# Find the examples where the 'medal' variable is not defined (NaN)
# Use the tilde (~) to get the inverse
# Create a filter to identify the examples where the 'medal' variable is defined

medal_column_filter = ~pd.isnull(medal_column)
```

```python
data = data[medal_column_filter]
data.head()
data.info()
```

```python
# Save the cleaned dataset to a new CSV file
data.to_csv('cleaned_olympics_data.csv', index=False)
```

```python

def plot_histogram(variable):
    """
    Input: Column name (variable)
    Output: Histogram plot of the specified variable
    """
    plt.figure(figsize=(5,3))
    plt.hist(data[variable], bins=80, color="#A3BDE6")
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title(f"Data Frequency - {variable.upper()}")
    plt.show()
```

```python
def plot_bar(data, column, top_n=5, color="#954DC7", figsize=(8, 5)):
    """
    Plots a bar chart showing the frequencies of the top N most common values 
    in a given categorical column of the dataset.

    Parameters:
    - data (pd.DataFrame): The dataset (DataFrame) containing the column.
    - column (str): The name of the categorical column for which the bar chart will be plotted.
    - top_n (int, optional): The number of top unique values to display (default is 5).
    - color (str, optional): The color for the bars in the chart (default is "#954DC7").
    - figsize (tuple, optional): The size of the figure (default is (8, 5)).

    Returns:
    - None: Displays the bar chart and prints the frequency counts of the top N values.
    """
    
    # Check if the column exists in the DataFrame
    if column not in data.columns:
        print(f"Error: Column '{column}' not found in the dataset.")
        return

    # Get the frequency count of the unique values in the specified column
    value_counts = data[column].value_counts().head(top_n)

    # Plot the bar chart
    plt.figure(figsize=figsize)
    plt.bar(value_counts.index, value_counts, color=color)
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.ylabel("Frequency")
    plt.title(f"Frequency Distribution - {column}")
    plt.tight_layout()  # Adjust layout to avoid overlap
    plt.show()

    # Print the frequency counts of the top N values
    print(f"Top {top_n} values in '{column}':\n{value_counts}\n")
```

```python
categorical_columns = ["name", "gender", "team", "noc", "season", "city", "sport", "event", "medal"]
for column in categorical_columns:
    plot_bar(data, column, top_n=5, color="#954DC7")
```

## Anomaly Detection

An **outlier** is a data point that is significantly different from the other observations in the dataset. An outlier can be caused by variability in the measurement or could indicate an experimental error; in the latter case, it is sometimes removed from the dataset. 

Outliers can cause serious problems in statistical analyses, leading to distorted conclusions. Therefore, detecting and handling outliers is an essential part of data preprocessing. Techniques like the **Z-score**, **IQR (Interquartile Range)**, and **visual methods (e.g., box plots)** can be used to identify outliers in the dataset.


```python
def anomali_detect(df, features):
    outlier_indices = []
    
    for feature in features:
        # 1st quartile
        Q1 = np.percentile(df[feature], 25)
        # 3rd quartile
        Q3 = np.percentile(df[feature], 75)
        # Interquartile Range (IQR)
        IQR = Q3 - Q1
        # Outlier detection step (1.5 times the IQR)
        outlier_step = IQR * 1.5
        # Detect outliers: those below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
        outlier_indices_for_feature = df[(df[feature] < (Q1 - outlier_step)) | (df[feature] > (Q3 + outlier_step))].index
        # Collect the outlier indices
        outlier_indices.extend(outlier_indices_for_feature)
    
    # Identify unique outlier indices and count their occurrences
    outlier_indices_count = Counter(outlier_indices)
    # Select samples that have outliers in more than one feature
    multiple_outliers = [index for index, count in outlier_indices_count.items() if count > 1]  # Changed 2 to 1
    
    return multiple_outliers

# Example usage
data_anomali = data.loc[anomali_detect(data, ["age", "height", "weight"])]
data_anomali.sport.value_counts()
```

```python
data_anomaly  = data.loc[anomali_detect(data,["age","height","weight"])]
data_anomaly.age.value_counts()
```

#### The sports with the most anomalies are shown above.


```python
plt.figure()
plt.bar(data_anomaly.sport.value_counts().index, data_anomaly.sport.value_counts().values, color="#D0F0C0")
plt.xticks(rotation=90)
plt.title("Sports with Anomalies")
plt.ylabel("Frequency")
plt.grid(True, alpha=0.3, color="#9ac884")
plt.show()
```

## Time Series Data Analysis

### Average Age, Height, and Weight Changes Over the Years
- Analyze the changes in **average age**, **height**, and **weight** over time.

### Medal Counts Over the Years
- Explore the **number of medals** won across different years.


#### 1. Average Age, Height, and Weight Changes Over the Years

```python
# Group the data by year and calculate the average of age, height, and weight for each year
avg_stats_by_year = data.groupby("year")[["age", "height", "weight"]].mean()
plt.figure(figsize=(10, 6))

# Plot average age over the years
plt.plot(avg_stats_by_year.index, avg_stats_by_year['age'], label='Average Age', color='blue', marker='o')

# Plot average height over the years
plt.plot(avg_stats_by_year.index, avg_stats_by_year['height'], label='Average Height (cm)', color='green', marker='s')

# Plot average weight over the years
plt.plot(avg_stats_by_year.index, avg_stats_by_year['weight'], label='Average Weight (kg)', color='red', marker='^')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Average Age, Height, and Weight Changes Over the Years')

# Add legend to distinguish between the lines
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

#### 2. Medal Counts Over the Years

```python
# Group the data by year and count the number of medals for each year
medal_counts_by_year = data.groupby("year")["medal"].value_counts().unstack(fill_value=0)
plt.figure(figsize=(10, 6))
medal_counts_by_year.plot(kind='bar', stacked=True, color=['gold', 'silver', 'brown'], width=0.8)
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.title('Number of Medals Won Over the Years')
plt.grid(True, axis='y', alpha=0.3)
plt.show()
```
